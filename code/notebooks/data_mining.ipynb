# read labels file for data extraction

```python
import pandas as pd
```

```python
labels = pd.read_csv('labels/egypt.csv', converters={i: str for i in range(100)})
```

```python
records = {}
page = labels['page'].iloc[0]

for p in range(len(labels.index)):
    for l in labels[labels.columns[2:]].iloc[p]:
        sparql.setQuery( """ PREFIX dc: <http://purl.org/dc/elements/1.1/>
        PREFIX edm: <http://www.europeana.eu/schemas/edm/>
        PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
        PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

        SELECT ?identifier ?schema ?value
        WHERE {{ ?identifier dc:identifier 'Call Nr:APM{}'.
          ?identifier ?schema ?value.
                  }} """ .format(l) )

        try:
            ret = sparql.queryAndConvert()
            record = ''

            for r in ret["results"]["bindings"]:
                if r['value']['type'] == 'literal':
                    record += r['value']['value'] + ', '

            if page in records:
                records[l] += [record]
            else: 
                records[l] = [record]

        except Exception as e:
            print(e)
```

```python
r = pd.DataFrame.from_dict(records, orient='index')
```
